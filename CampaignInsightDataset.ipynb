```python
import pandas as pd #work with table
import numpy as np #generate random values like number and category

# set a random seed
np.random.seed(42)
```

```python
dates = pd.date_range("2026-01-01", "2026-06-30")

channels = ["Search", "Social", "Display", "Native"]
segments = ["First Party", "Retargeting", "Third Party", "Lookalike", "Platform Native Targeting"]

# Define the number of rows (campaigns) first
num_rows = 200
campaign_ids = {10000 + i for i in range(num_rows)} # Corrected syntax for set comprehension

channel_params = {
    "Search": {"ctr": 0.05, "cpc": 2.5},
    "Social": {"ctr": 0.03, "cpc": 1.5},
    "Display": {"ctr": 0.01, "cpc": 0.8},
    "Native": {"ctr": 0.08, "cpc": 0.3}
}

segment_cr = {
    "First Party": 0.20,
    "Retargeting": 0.10,
    "Third Party": 0.06,
    "Lookalike": 0.04,
    "Platform Native Targeting": 0.02
}

# Initialize an empty list to collect the generated data rows
all_data_rows = []

for date in dates:
    for channel in channels:
        for segment in segments:
          for campaign_id in campaign_ids:

            impressions = np.random.randint(5000, 50000)

            ctr = channel_params[channel]["ctr"] * \
                  np.random.uniform(0.8, 1.2)

            clicks = int(impressions * ctr)

            cpc = channel_params[channel]["cpc"] * \
                  np.random.uniform(0.9, 1.1)

            spend = round(clicks * cpc, 2)

            conv_rate = segment_cr[segment] * \
                        np.random.uniform(0.8, 1.2)

            conversions = int(clicks * conv_rate)

            revenue = round(
                conversions *
                np.random.uniform(50, 250),
                2
            )

            all_data_rows.append([
                date,
                campaign_id,
                channel,
                segment,
                impressions,
                clicks,
                spend,
                conversions,
                revenue
            ])

df = pd.DataFrame(
    all_data_rows,
    columns=[
        "date",
        "campaign_id",
        "channel",
        "audience_segment",
        "impressions",
        "clicks",
        "spend",
        "conversions",
        "revenue"
    ]
)

df["ctr"] = df["clicks"] / df["impressions"]
df["cpc"] = df["spend"] / df["clicks"]
df["conversion_rate"] = (
    df["conversions"] / df["clicks"]
)

print(df.head())

df.to_csv("campaign_data.csv", index=False)
```
